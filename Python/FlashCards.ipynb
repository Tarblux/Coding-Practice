# Setting Up Neighbors Helper Functions

```python
# For grids where the neighbor needs to be 1 / a specific value

class Solution:

    def __init__(self):
        self.directions = [(-1,0),(1,0),(0,-1),(0,1)]

    def neighbors(self,matrix,visited,r,c):

        for dr,dc in self.directions:
            nr = r + dr
            nc = c + dc

            if nr < 0 or nr >= len(matrix) or nc < 0 or nc >= len(matrix[0]):
                continue

            if (nr,nc) in visited or matrix[nr][nc] != 1:
                continue

            yield(nr,nc)
```

```python
# Open the Lock with Generator Pattern (We need to generate code increases/decreases from the given code)
# Main thing to note here is how we use the current string to generate the next neighbors

from typing import List
from collections import deque

class Solution:
    
    def neighbors(self, code: str, moves: int):
        for i in range(len(code)):
            up = (int(code[i]) + 1) % 10
            down = (int(code[i]) - 1) % 10
            
            up_string = code[:i] + str(up) + code[i+1:]
            down_string = code[:i] + str(down) + code[i+1:]
            
            yield (up_string, moves + 1)
            yield (down_string, moves + 1)
    
    def openLock(self, deadends: List[str], target: str) -> int:
        visited = set()
        queue = deque([('0000', 0)])
        
        for de in deadends:
            visited.add(de)
        
        while queue:
            state, moves = queue.popleft()
            
            if state in visited:
                continue
            
            visited.add(state)
            
            if state == target:
                return moves
            
            for neighbor, neighbor_moves in self.neighbors(state, moves):
                if neighbor not in visited:
                    
                    queue.append((neighbor, neighbor_moves))

        return -1

# Example usage
test = Solution()
deadends = ["0201", "0101", "0102", "1212", "2002"]
target = "0203"
print(test.openLock(deadends, target))  # Output should be the number of moves to unlock
```

```python
# 127. Word Ladder Question , Neighbors are words with a one character difference

class Solution:

    def validNeighbors(self,wordList,visited,word):

        for i in range(len(word)):
            for j in range(97,123):

                new_word = word[:i] + chr(j) + word[i+1:]

                if new_word not in wordList or new_word in visited:
                    continue

                yield new_word
```
